```python
import pandas
import matplotlib.pyplot
import seaborn
import datetime
```

```python
tips_data = pandas.read_csv('tips.csv')
tips_data
```

```python
tips_data.shape
```

```python
tips_data.describe()
```

```python
tips_data.groupby(by = 'time').mean()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.barplot(x = 'time', y = 'total_bill', data = tips_data)
matplotlib.pyplot.title('Bills and Tips')
matplotlib.pyplot.xticks(rotation = 90)
matplotlib.pyplot.show()
```

```python
mean = tips_data['tip'].mean()
mean
```

```python
tips_data['Above Average'] = (tips_data['tip'] - mean) > 0
tips_data[['tip', 'Above Average']]
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.countplot('time', hue = 'Above Average',
                 data = tips_data,
                  order = tips_data['time'].value_counts().index)
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.countplot('time', hue = 'sex',
                 data = tips_data,
                  order = tips_data['time'].value_counts().index)
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.boxplot(x = 'sex', y = 'total_bill',
               data = tips_data, palette = 'nipy_spectral')
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.boxplot(x = 'size', y = 'total_bill',
               data = tips_data, palette = 'nipy_spectral')
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
seaborn.catplot(x = 'size', y = 'total_bill',
               hue = 'time', data = tips_data,
               kind = 'bar', height = 6,
               aspect = 2, palette = 'CMRmap')
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.stripplot(x = 'size', y = 'tip',
                 data = tips_data)
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
seaborn.swarmplot(x = 'size', y = 'tip',
                 data = tips_data)
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.swarmplot(x = 'size', y = 'total_bill',
                 hue = 'Above Average', data = tips_data)
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.swarmplot(x = 'size', y = 'total_bill',
                 hue = 'Above Average', data = tips_data, dodge = True)
matplotlib.pyplot.title('Bill and Tips')
matplotlib.pyplot.show()
```

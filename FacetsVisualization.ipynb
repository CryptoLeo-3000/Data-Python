```python
import pandas
import seaborn
import matplotlib
import numpy
```

```python
automobile_data_processed = pandas.read_csv('processed_automobile_data.csv')
automobile_data_processed
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
fg = seaborn.FacetGrid(automobile_data_processed,
                      col = 'fuel-type',
                      height = 7, aspect = 1)
fg.map(matplotlib.pyplot.hist, 'price', color = 'red')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
fg = seaborn.FacetGrid(automobile_data_processed,
                      col = 'fuel-type',
                      height = 7, aspect = 1)
fg.map(seaborn.scatterplot, 'highway-mpg', 'price', s = 100)
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
fg = seaborn.FacetGrid(automobile_data_processed,
                      col = 'fuel-type',
                       hue = 'num-of-cylinders',
                      height = 7, aspect = 1)
fg.map(seaborn.scatterplot, 'highway-mpg', 'price', s = 100)
fg.add_legend()
matplotlib.pyplot.show()
```

```python
tips_data = pandas.read_csv('tips.csv')
tips_data
```

```python
matplotlib.pyplot.figure(figsize = (15, 10))
fg = seaborn.FacetGrid(tips_data, col = 'size',
                       height = 7, aspect = 1)
fg.map(matplotlib.pyplot.hist, 'total_bill', color = 'green')
matplotlib.pyplot.show()
```

```python
group_size_values = numpy.sort(tips_data['size'].unique())
```

```python
fg = seaborn.FacetGrid(tips_data, row = 'time',
                      height = 6, aspect = 1)
fg.map(seaborn.swarmplot, 'size',
      'total_bill', order = group_size_values)
fg.add_legend()
```

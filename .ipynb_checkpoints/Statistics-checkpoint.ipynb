```python
import pandas
import matplotlib

crude_oil = pandas.read_csv("Processed_U.S._Crude_Oil.csv")
crude_oil
```

```python
crude_oil.plot(x = 'Months', y = 'U.S. Crude Oil ', figsize = (12, 8), color = 'green')
matplotlib.pyplot.xlabel("Dates")
matplotlib.pyplot.ylabel("Productions")
matplotlib.pyplot.title("U.S. Crude Oil Production")
```

```python
crude_oil.boxplot('U.S. Crude Oil ', figsize = (12, 8))
matplotlib.pyplot.ylabel("Productions")
matplotlib.pyplot.title("U.S. Crude Oil Production")
```

```python
crude_oil.boxplot('California', figsize = (12, 8))
matplotlib.pyplot.ylabel("Productions")
matplotlib.pyplot.title("California Production")
```

```python
crude_oil[['Alaska', 'California']].boxplot(figsize = (12, 8))
matplotlib.pyplot.ylabel("Productions")
matplotlib.pyplot.title("Alaska and California")
```

```python
crude_oil.boxplot(column = ['U.S. Crude Oil '], by = ['Year'], figsize = (12, 8))
matplotlib.pyplot.ylabel("Productions")
matplotlib.pyplot.title("U.S. Crude Oil Production")
```

```python
year_data = crude_oil.groupby('Year', as_index = False).sum()
year_data[['Year', 'U.S. Crude Oil ']]
```

```python
year_data.plot.bar(x = 'Year', y = 'U.S. Crude Oil ', figsize = (12, 8), color = 'green')
matplotlib.pyplot.xlabel("Year")
matplotlib.pyplot.ylabel("Production")
matplotlib.pyplot.title("U.S. Crude Oil Production")
```

```python
mean_production = crude_oil.mean()[2:-3]
mean_production.sort_values(ascending = False)
```

```python
mean_prod_data = pandas.DataFrame(mean_production).reset_index()
mean_prod_data.columns = ['State', 'Production']
mean_prod_data
```

```python
mean_production.sort_values(ascending = False).plot.bar(x = 'State', y = 'Production', figsize = (12, 8))
matplotlib.pyplot.xlabel('State')
matplotlib.pyplot.ylabel('Production')
matplotlib.pyplot.title('State production Graph')
```

```python
mean_prod_data['Production'].plot.kde(figsize = (12, 8))
matplotlib.pyplot.title('U.S. Oil Production of Individual States')
matplotlib.pyplot.xlabel('Oil Production')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
matplotlib.pyplot.scatter(crude_oil['Texas'], crude_oil['U.S. Crude Oil '], color= 'green')
matplotlib.pyplot.xlabel('U.S. Production')
matplotlib.pyplot.ylabel('Texas Poduction')
matplotlib.pyplot.show()
```

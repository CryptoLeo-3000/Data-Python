```python
import pandas
import seaborn
import matplotlib.pyplot
```

```python
Automobile_data = pandas.read_csv("processed_automobile_data.csv")
Automobile_data
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.distplot(Automobile_data['price'], color = 'red')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.distplot(Automobile_data['price'], bins = 20, color = 'red')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.distplot(Automobile_data['price'], hist = False, color = 'red')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.distplot(Automobile_data['price'], hist = False, 
                 rug = True, color = 'red')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.rugplot(Automobile_data['price'], height = 0.5, color = 'red')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.kdeplot(Automobile_data['price'], shade = True, color = 'red')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.scatterplot(x = 'horsepower', y = 'price',
                    data = Automobile_data, s = 120)
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.scatterplot(x = 'horsepower', y = 'price',
                    data = Automobile_data, hue = 'num-of-cylinders',
                    s = 120)
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.regplot(x = 'horsepower', y = 'price',
                    data = Automobile_data)
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.regplot(x = 'highway-mpg', y = 'price',
                    data = Automobile_data)
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.jointplot(x = 'horsepower', y = 'price',
                    data = Automobile_data)
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.jointplot(x = 'horsepower', y = 'price',
                    data = Automobile_data, kind = 'reg')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.jointplot(x = 'horsepower', y = 'price',
                    data = Automobile_data, kind = 'kde')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize = (12, 8))
seaborn.jointplot(x = 'horsepower', y = 'price',
                    data = Automobile_data, kind = 'hex')
matplotlib.pyplot.title("Automobile Data")
matplotlib.pyplot.show()
```

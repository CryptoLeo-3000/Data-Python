```python
import numpy
import pandas
import matplotlib.pyplot
```

```python
bikesharing_hour = pandas.read_csv('bike_sharing_hourly.csv')
bikesharing_hour
```

```python
bikesharing_hour[['temp', 'hum']].describe()
```

```python
bikesharing_hour[['temp', 'hum']].corr()
```

```python
bikesharing_hour['temp'].autocorr(lag = 2)
```

```python
bikesharing_hour['temp'].autocorr(lag = 12)
```

```python
fig, (ax1, ax2) = matplotlib.pyplot.subplots(2, 1, figsize = (12, 8))
ax1.acorr(bikesharing_hour['temp'],maxlags = 12, color = 'green')
ax1.title.set_text('Temperature')
ax1.set_xlabel('Lags', fontsize = 15)

ax2.acorr(bikesharing_hour['hum'], maxlags = 12, color = 'orange')
ax2.title.set_text('Humidity')
ax2.set_xlabel('Lags', fontsize = 15)
matplotlib.pyplot.show()
```

```python
fig, (ax1, ax2) = matplotlib.pyplot.subplots(1, 2, figsize = (12, 8))
ax1.acorr(bikesharing_hour['temp'],maxlags = 24, color = 'green')
ax1.title.set_text('Temperature')
ax1.set_xlabel('Lags', fontsize = 15)

ax2.acorr(bikesharing_hour['hum'], maxlags = 24, color = 'orange')
ax2.title.set_text('Humidity')
ax2.set_xlabel('Lags', fontsize = 15)

matplotlib.pyplot.suptitle('Autocorrelation')
matplotlib.pyplot.show()
```
